### 接線の描画
導関数を使って、関数$f(x)=3x^2+4x-5$の$x=1$における接線を描画します。

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return 3*x**2 + 4*x - 5

def my_func_dif(x):  # 導関数
    return 6*x + 4

x = np.linspace(-3, 3)
y = my_func(x)

a = 1
y_t = my_func_dif(a)*x + my_func(a) - my_func_dif(a)*a  # x=1のときの接線

plt.plot(x, y, label="y")
plt.plot(x, y_t, label="y_t")

plt.legend()
plt.show()
```

### ネイピア数の実装
ネイピア数は、NumPyにおいて`e`で取得することができます。  
また、ネイピア数の累乗はNumPyのexp関数で実装することができます。

```python
import numpy as np

print(np.e)  # ネイピア数
print(np.exp(1))  #eの1乗
```

### 自然対数の実装
自然対数は、NumPyのlog関数を使って実装することができます。

```python
import numpy as np

print(np.log(np.e))  # eはeの1乗
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 2)
y_e = np.exp(x)  # ネイピア数の累乗
y_log = np.log(x)  # 自然対数

plt.plot(x, y_e, label="y_e")
plt.legend()
plt.show()

plt.plot(x, y_log, label="y_t")
plt.legend()
plt.show()
```
